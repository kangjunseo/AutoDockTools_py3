```python
#      
def is_abnormal_pdbqt(file_path):
    """
    PDBQT   ATOM  X, Y, Z   0   

    Args:
    file_path (str): PDBQT  

    Returns:
    bool:    
    """
    with open(file_path, 'r') as file:
        lines = file.readlines()
        
    atom_lines = [line for line in lines if line.startswith('ATOM')]
    
    for line in atom_lines:
        parts = line.split()
        print(parts)
        try:
            x, y, z = float(parts[8]), float(parts[6]), float(parts[7])
            print(x,y,z)
            if x == 0.0 and y == 0.0 and z == 0.0:
                return True
        except (IndexError, ValueError):
            continue
    
    return False

#   
file_path = 'DP/pdbqt/isoleucinyltryptophan.pdbqt'
is_abnormal = is_abnormal_pdbqt(file_path)
print(f"The file is {'abnormal' if is_abnormal else 'normal'}.")
```

```python
import matplotlib.pyplot as plt

def draw_hist(targets):
    #targets : must be 2 row - list
    #target[0] should be data, target[1] should be label
    plt.figure(figsize=(10, 6))
    for target in targets:
        plt.hist(target[0], bins=20, edgecolor='k', alpha=0.7, label=target[1])
    plt.title('Distribution of Affinity Values')
    plt.xlabel('Affinity')
    plt.ylabel('Counts')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
def dock_dir(receptor, pdbqt_dir, output_dir):
    import os, time
    import dock
    os.makedirs(output_dir, exist_ok=True)
    cnt = 0

    total_start = time.time()
    for dirpath, dirnames, filenames in os.walk(pdbqt_dir):
        for file in filenames:
            if file.endswith('.pdbqt'):
                cnt+=1
                print(f'##########{cnt}##########{cnt}##########{cnt}##########')
                start_time = time.time()
                dock.dock_ligand(receptor, os.path.join(pdbqt_dir, file) , os.path.join(output_dir, f'docked_{file[:-6]}.pdbqt'))
                end_time = time.time()
                print(end_time - start_time)
    total_end = time.time()
    print(total_end - total_start)
```

```python
receptor = 'receptor/apo_8gfa.pdbqt'
receptor2 = 'receptor/apo8gfa_cap_f.pdbqt'
pdbqt_dir = 'DP/pdbqt/'
output_dir = 'DP/apo_dock2/'
output_dir2 = 'DP/cap_dock2/'

dock_dir(receptor, pdbqt_dir, output_dir)
dock_dir(receptor2, pdbqt_dir, output_dir2)
```
